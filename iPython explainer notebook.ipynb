## Basic statistics

Initially the dataset consisted of 4.96 million data observations, all symbolising one a incidents where San Francisco Fire Department received a call for service. These responses where made up of 32 different call types ranging from a medical incident, traffic collision, citizens assist etc. If was therefore decided to narrow down the call types to only include incidents which included a type of fire. Therefore, the dataset was filteret to include the call types:

Structure Fire,Vehicle Fire, Outdoor Fire,
Marine Fire, Smoke Investigation, Explosions and Train/Rail Fire.

The reason for this filtering was primarily due the fact that we found these types more interesting to work with due to the fact that buildings in San Francisco are placed very tight to eachother and the consequences thereby are a lot greater, which makes the response time of the fire department al desto more important. 

After narrowing down the dataset, basic cleansing procedures were done to the dataset. For instance NaN-values were removed in the columns that we needed to calculate response times and data from 2019 was removed as well. After doing so 450.000 observations were left. 

Each row representing an incident, where we had: Call type, neighbourhood, year, month, Battalion, location, time Recieved call, time on scene.

## Genre

### Genre:

The genre chosen for this project is a combination of the annotated chart and (interactive) slide show. These were chosen to give the user the best possible opportunities to invetigate as much as possible, since the main belief in this porject is that telling a story through data should be interactive and the user should be able to explore as much as possible on their own in an intuitive way - by clicking, sliding, scrolling and hovering on the charts.   
    
### Visual Narrative:
From the visual narrative we chose to incorporate the tools: Progress bar/timebar, motion and zoom functions. 

The progress bar/timebar is included twice in the data viz. It allows for the user to slide over the different years provided and to adjust how zoomed in the user wants to be on the scatter plot. The reason for chosen this type of bar, was to give the user a explorative option to slide over the different years and thereby se how the connecting bar charts change interactively.

Initially a progress bar/timebar was also chosen in the map illustration of San Francisco, to give the user control over which time of the day the wanted to investigate. However, it seemed more interesting to change this function such that this was done automatically by a generator such that every time of the day in the month was investigated by watching a animated plot. The generator timecontroller was created to run over each hour of the day, repeating itself when done. 

Furthermore, we have for instance added a zoom function on the scatter plot since the best stories sometimes are told in details. 

### Narrative Structure:
The narrative structure of our story has primarily been focused around the interactivity, where hovering and filtering are the most important ones, since these are what initially will be displayed on the final plots, where some of the most interesting insights can be found. In addition the aim was to make the plots as intuitive as possible but it was found necessary to add some summaries and messages in order to ensure that the user would utilize the full potential of the plots.
